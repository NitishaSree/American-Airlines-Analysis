# Data Cleaning and Data Preprocessing

```python
from scipy import stats


# Function for data cleaning and preprocessing
def clean_and_preprocess(df):
    # Check for missing values
    print("Missing values per column:\n", df.isnull().sum())

    # Drop rows with missing values in essential columns
    essential_columns = ['passengers', 'carrier_lg', 'city1', 'city2']
    df.dropna(subset=[col for col in essential_columns if col in df.columns], inplace=True)

    # Remove rows with zero or negative passengers (assuming these are invalid)
    if 'passengers' in df.columns:
        df = df[df['passengers'] > 0]
    
    # Fill missing values for numeric columns with appropriate statistics
    fill_values = {}
    if 'nsmiles' in df.columns:
        fill_values['nsmiles'] = df['nsmiles'].mean()
    if 'passengers' in df.columns:
        fill_values['passengers'] = df['passengers'].median()
    df.fillna(value=fill_values, inplace=True)

    # Remove duplicates
    print("Removing duplicates...")
    df.drop_duplicates(inplace=True)

    # Overview of numerical features
    print("Numerical features overview:\n", df.describe())
    
    # Overview of categorical features
    print("Categorical features overview:\n", df.describe(include=['object']))

    # Identify outliers using Z-scores for numerical columns
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    if len(numeric_columns) > 0:
        z_scores = stats.zscore(df[numeric_columns])
        outlier_condition = (np.abs(z_scores) > 3).any(axis=1)
        df_outliers = df[outlier_condition]
        print(f'Number of outliers detected: {len(df_outliers)}')

        # Visualize outliers using box plots for each numeric column
        for column in numeric_columns:
            plt.figure(figsize=(8, 4))
            sns.boxplot(y=df[column], palette='crest')
            plt.title(f'Box Plot of {column}', fontsize=16)
            plt.ylabel(column, fontsize=12)
            plt.show()

    return df

# Load datasets
fuel_rates = pd.read_excel('Quarterly Fuel rates.xlsx')
admin_mail = pd.read_excel('GeneralAdmin_Mail.xlsx')
merged_data = pd.read_excel('merged_excel_file.xlsx')
cleaned_dataset = pd.read_csv("cleaned_dataset.csv")

# Apply cleaning and preprocessing to each dataset and save results
print("\nProcessing fuel_rates dataset:")
cleaned_fuel_rates = clean_and_preprocess(fuel_rates)

print("\nProcessing admin_mail dataset:")
cleaned_admin_mail = clean_and_preprocess(admin_mail)

print("\nProcessing merged_data dataset:")
cleaned_merged_data = clean_and_preprocess(merged_data)

print("\nProcessing cleaned_dataset:")
cleaned_cleaned_dataset = clean_and_preprocess(cleaned_dataset)


```

# Market Share Analysis Before and After COVID

```python
# Split data into pre-COVID and post-COVID periods
pre_covid = cleaned_dataset[(cleaned_dataset['Year'] >= 2015) & (cleaned_dataset['Year'] < 2020)]
post_covid = cleaned_dataset[(cleaned_dataset['Year'] >= 2020) & (cleaned_dataset['Year'] <= 2024)]

# Market share before COVID
pre_covid_marketshare = pre_covid.groupby('carrier_lg')['passengers'].sum() / pre_covid['passengers'].sum()

# Market share after COVID
post_covid_marketshare = post_covid.groupby('carrier_lg')['passengers'].sum() / post_covid['passengers'].sum()

# Visualize market share shifts
# Pre-COVID market share plot with annotations
plt.figure(figsize=(10, 6))
sns.barplot(x=pre_covid_marketshare.index, y=pre_covid_marketshare.values, color='blue', alpha=0.7)
plt.title('Market Share Before COVID (2015-2020)')
plt.xlabel('Airline')
plt.ylabel('Percentage Market Share')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Adding annotations
for i in range(len(pre_covid_marketshare)):
    plt.text(i, pre_covid_marketshare.values[i] + 0.002, 
             f'{pre_covid_marketshare.values[i]*100:.2f}%', 
             ha='center', fontsize=10)

plt.show()

# Post-COVID market share plot with annotations
plt.figure(figsize=(10, 6))
sns.barplot(x=post_covid_marketshare.index, y=post_covid_marketshare.values, color='orange', alpha=0.7)
plt.title('Market Share After COVID (2020-2024)')
plt.xlabel('Airline')
plt.ylabel('Percentage Market Share')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Adding annotations
for i in range(len(post_covid_marketshare)):
    plt.text(i, post_covid_marketshare.values[i] + 0.002, 
             f'{post_covid_marketshare.values[i]*100:.2f}%', 
             ha='center', fontsize=10)

plt.show()
```

```python

# Calculate market share percentage change
market_share_change = (post_covid_marketshare - pre_covid_marketshare) * 100

# Remove any rows with non-finite (NaN or inf) values
market_share_change = market_share_change[np.isfinite(market_share_change)]

# Visualize the change in market share
plt.figure(figsize=(10, 6))
sns.barplot(x=market_share_change.index, y=market_share_change.values, hue=market_share_change.index, palette='coolwarm', alpha=0.7, dodge=False)
plt.title('Market Share Change After COVID (2020-2024 vs 2015-2020)')
plt.xlabel('Airline')
plt.ylabel('Market Share Change (%)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Adding annotations on top of bars
for i in range(len(market_share_change)):
    plt.text(i, market_share_change.values[i] + 0.1, 
             f'{market_share_change.values[i]:.2f}%', 
             ha='center', fontsize=10)

plt.show()

```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Plot the relationship between 'passengers' and market share ('large_ms' as an example)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cleaned_dataset, x='passengers', y='large_ms')
plt.title('Passengers vs Market Share (large_ms)', fontsize=15)
plt.xlabel('Passengers')
plt.ylabel('Market Share (large_ms)')
plt.show()

# Plot the relationship between 'fare' and market share ('large_ms' as an example)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cleaned_dataset, x='fare', y='large_ms')
plt.title('Fare vs Market Share (large_ms)', fontsize=15)
plt.xlabel('Fare')
plt.ylabel('Market Share (large_ms)')
plt.show()

# Correlation analysis: Check the correlation between variables
correlation = cleaned_dataset[['passengers', 'fare', 'large_ms', 'lf_ms']].corr()
print(correlation)

# Regression analysis to see the effect of 'passengers' and 'fare' on 'large_ms'
# Defining the independent variables (X) and dependent variable (Y)
X = cleaned_dataset[['passengers', 'fare']]  # Independent variables
X = sm.add_constant(X)  # Adding constant for intercept
y = cleaned_dataset['large_ms']  # Dependent variable (market share)

# Fit the model
model = sm.OLS(y, X).fit()

# Print out the regression results
print(model.summary())

# Optional: Check for the other market share, 'lf_ms'
X2 = cleaned_dataset[['passengers', 'fare']]
X2 = sm.add_constant(X2)
y2 = cleaned_dataset['lf_ms']

# Fit the model
model2 = sm.OLS(y2, X2).fit()

# Print out the regression results for 'lf_ms'
print(model2.summary())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Calculate average fare and total passengers for pre-COVID and post-COVID periods

# Average fare and total passengers before COVID
pre_covid_fare_passengers = pre_covid.groupby('carrier_lg').agg(
    average_fare=('fare', 'mean'),
    total_passengers=('passengers', 'sum')
)

# Average fare and total passengers after COVID
post_covid_fare_passengers = post_covid.groupby('carrier_lg').agg(
    average_fare=('fare', 'mean'),
    total_passengers=('passengers', 'sum')
)

# Step 2: Visualize average fare before and after COVID

# Plotting average fare before and after COVID
plt.figure(figsize=(12, 6))

# Plot pre-COVID average fare
plt.subplot(1, 2, 1)
sns.barplot(x=pre_covid_fare_passengers.index, y=pre_covid_fare_passengers['average_fare'], color='blue', alpha=0.7)
plt.title('Average Fare Before COVID (2015-2020)')
plt.xlabel('Airline')
plt.ylabel('Average Fare')
plt.xticks(rotation=45)

# Plot post-COVID average fare
plt.subplot(1, 2, 2)
sns.barplot(x=post_covid_fare_passengers.index, y=post_covid_fare_passengers['average_fare'], color='orange', alpha=0.7)
plt.title('Average Fare After COVID (2020-2024)')
plt.xlabel('Airline')
plt.ylabel('Average Fare')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Step 3: Visualize total passengers before and after COVID

# Plotting total passengers before and after COVID
plt.figure(figsize=(12, 6))

# Plot pre-COVID total passengers
plt.subplot(1, 2, 1)
sns.barplot(x=pre_covid_fare_passengers.index, y=pre_covid_fare_passengers['total_passengers'], color='blue', alpha=0.7)
plt.title('Total Passengers Before COVID (2015-2020)')
plt.xlabel('Airline')
plt.ylabel('Total Passengers')
plt.xticks(rotation=45)

# Plot post-COVID total passengers
plt.subplot(1, 2, 2)
sns.barplot(x=post_covid_fare_passengers.index, y=post_covid_fare_passengers['total_passengers'], color='orange', alpha=0.7)
plt.title('Total Passengers After COVID (2020-2024)')
plt.xlabel('Airline')
plt.ylabel('Total Passengers')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Step 4: Calculate the percentage change in average fare and total passengers

# Calculate percentage change in average fare and total passengers
fare_change = (post_covid_fare_passengers['average_fare'] - pre_covid_fare_passengers['average_fare']) / pre_covid_fare_passengers['average_fare'] * 100
passenger_change = (post_covid_fare_passengers['total_passengers'] - pre_covid_fare_passengers['total_passengers']) / pre_covid_fare_passengers['total_passengers'] * 100

# Step 5: Visualize the change in fare and passengers

# Plotting fare change
plt.figure(figsize=(10, 6))
sns.barplot(x=fare_change.index, y=fare_change.values, palette='coolwarm', alpha=0.7)
plt.title('Percentage Change in Average Fare After COVID')
plt.xlabel('Airline')
plt.ylabel('Percentage Change in Average Fare (%)')
plt.xticks(rotation=45)

# Adding annotations
for i in range(len(fare_change)):
    plt.text(i, fare_change.values[i] + 0.2, 
             f'{fare_change.values[i]:.2f}%', 
             ha='center', fontsize=10)

plt.tight_layout()
plt.show()

# Plotting passenger change
plt.figure(figsize=(10, 6))
sns.barplot(x=passenger_change.index, y=passenger_change.values, palette='coolwarm', alpha=0.7)
plt.title('Percentage Change in Total Passengers After COVID')
plt.xlabel('Airline')
plt.ylabel('Percentage Change in Total Passengers (%)')
plt.xticks(rotation=45)

# Adding annotations
for i in range(len(passenger_change)):
    plt.text(i, passenger_change.values[i] + 0.2, 
             f'{passenger_change.values[i]:.2f}%', 
             ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Ensure the 'cleaned_dataset_filtered' is properly defined. If not, filter the dataset as needed.
# For example, if the original dataset is 'cleaned_dataset':
cleaned_dataset_filtered = cleaned_dataset[cleaned_dataset['Year'].between(2015, 2023)]  # Adjust the year range if needed

# Map carrier codes to full names if necessary
cleaned_dataset_filtered['carrier_lg'] = cleaned_dataset_filtered['carrier_lg'].replace({
    'AA': 'American Airlines', 
    'UA': 'United', 
    'DL': 'Delta', 
    'NK': 'Spirit', 
    'WN': 'Southwest', 
    'AS': 'Alaska'
})

# Calculate average fare for each carrier and year
top_carriers = ['American Airlines', 'United', 'Delta', 'Spirit', 'Southwest', 'Alaska']  
average_fare_carrier = cleaned_dataset_filtered[cleaned_dataset_filtered['carrier_lg'].isin(top_carriers)] \
    .groupby(['Year', 'carrier_lg'])['fare'].mean().reset_index()

# Plotting the annual average fare by carrier
plt.figure(figsize=(12, 6))
sns.barplot(data=average_fare_carrier, x='Year', y='fare', hue='carrier_lg', palette='pastel')
plt.title('Annual Average Fare by Major Carrier (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Average Fare ($)')
plt.legend(title='Carrier')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()  # To avoid clipping of labels
plt.show()
```
